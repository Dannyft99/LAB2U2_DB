```python
'''Importación de librerias'''
import pandas as pd
import numpy as np
import json
```

```python
'''Carga de dataset'''
filename = 'data_act_01.csv'
df = pd.read_csv(filename)
df
```

```python
'''Debido a que la fila Range no contiene valores agregados se eliminará el mismo. Ya que este no es relevante,
esto es facil de determinar con una imple impresion de la base de datos ya que se observa que no hay ningun valor
agregado.Para ello con el comando drop eliminaremos el mismo, esto con la finalidad de evitar problemas con datos
que no son necesarios.'''
data=df.drop(['Range'], axis=1)
data
```

```python
'''Comprobación de datos duplicados'''
data.duplicated().sum()
```

```python
'''Comprobación duplicidad en la columna CrimeId'''
data['CrimeId'].duplicated().sum()
```

```python
'''Identificación de registros de CrimeId Duplicados'''
duplicates = data[data['CrimeId'].duplicated(keep=False)]['CrimeId'].tolist()
duplicates 
```

```python
'''Funcion para realizar un incremento en el valor ID para asi cambiarlo y evidar redundancia'''
set_duplicates = set(duplicates)

for r in set_duplicates:
    dup = data[data['CrimeId'] == r]
    data.loc[dup.iloc[1,:].name, 'CrimeId'] = data.loc[dup.iloc[1,:].name]['CrimeId'] + 12
```

```python
'''Comprobación duplicidad en la columna CrimeId'''
data['CrimeId'].duplicated().sum()
```

# Localizar valores nulos

```python
'''Sumatoria de valores nulos'''
data.isnull().sum()
```

#### Aqui podemos observar que en la columna City y en la Columna Address existen valores nulos

```python
'''El problema detectado para la columna State es que existen valores nulos. Para lo cual como solucion despues de 
realizar el analisis del dataset, se puede observar que los registros tomados pertenecen al estado CA. Entonces,
se completaran los datos. Tener datos nulos en un dataset es incorrecto ya que se trata de registros incompletos'''
data = data[(data['State'] == 'CA')]
```

```python
'''El problema detectado para la columna City es que existen valores nulos. Para lo cual como solucion despues de 
realizar el analisis del dataset, se puede observar que los registros tomados pertenecen a la ciudad de 
San Francisco. Entonces, se completaran los datos. Tener datos nulos en un dataset es incorrecto ya que se
trata de registros incompletos'''
data = data[(data['City'] == 'San Francisco')]
```

```python
'''Verificación de datos nulos'''
data.isnull().sum()
```

```python
'''Con la finalidad de eliminar la redundancia de datos, se eliminaran los registros de OffenseDate y CallTime. 
Esta decisión fue tomada, debido a que existe un registro que contempla estos dos registros llamado CallDateTime.'''
data=data.drop(['OffenseDate','CallTime'], axis=1)
data
```
